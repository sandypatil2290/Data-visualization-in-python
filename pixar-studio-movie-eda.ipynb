**Movie with HIghest budget**

```python
plt.figure(figsize=(15,10))
sns.barplot(x="Budget(million)",y="Moviename",data=df.sort_values("Budget(million)",ascending=False),hue="Budget(million)",alpha=1);plt.xticks(rotation=90)
plt.title("Movie with highest Budget" );plt.show()
```

**Pixar studio movie profit in us and canada** 

```python
plt.figure(figsize=(15,10))
sns.barplot(x="Moviename",y="US and canada(million)",data=df.sort_values("US and canada(million)",ascending=False),hue="Moviename",alpha=1);plt.xticks(rotation=90)
plt.title("Movie with highest profit in us and canada" );plt.show()
```

**Pixar studio movie profit in Other territories** 

```python
plt.figure(figsize=(15,10))
sns.barplot(x="Moviename",y="Other territories(million)",data=df.sort_values("Other territories(million)",ascending=False),hue="Moviename",alpha=1);plt.xticks(rotation=90)
plt.title("Movie with highest profit in Other territories(million)" );plt.show()
```

**Pixar studio movie profit in worldwide**

```python
plt.figure(figsize=(15,10))
sns.barplot(x="Moviename",y="Worldwide(million)",data=df.sort_values("Worldwide(million)",ascending=False),hue="Moviename",alpha=1);plt.xticks(rotation=90)
plt.title("Movie with highest profit in Worldwide(million)" );plt.show()
```
